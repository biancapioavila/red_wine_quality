## Introdução

<P align = "justify" > A percepção da qualidade do vinho é um construto multidimensional que tem recebido especial interesse nos últimos anos, sendo entendido como uma combinação de fatores relacionados ao produto e ao consumidor. <BR>
<P align = "justify" > Quando o consumidor compra um produto, ele analisa características como marca, preço, embalagem, entre outros, para tomar sua decisão final. No entanto, o nível de influência que os atributos exercem sobre o consumidor varia de acordo com o produto e o perfil das pessoa. <BR>
<P align = "justify" > No caso do vinho, a variedade da uva, suas características físico-químicas, o país de origem, o teor alcoólico e o preço são alguns dos atributos do produto que podem ser levados em consideração no momento da compra. <BR>
<P align = "justify" > Com isso, no presente projeto, busca-se analisar a influência das características físico-químicas e sensoriais a fim de se predizer, um vinho tinto, que tenha atributos desejáveis.</P>

## Objetivos

No presente projeto, busca-se analisar a influência das características físico-químicas e sensoriais de vinho tinto, 
e assim obter modelos de marketing-alvo, gostos dos consumidores e nichos de mercado.

## Dados

<P align = "justify"> 1 - fixed acidity (acidez total) - medida da acidez devido à presença de ácidos orgânicos de baixa volatilidade (ácido málico, lático, tartárico ou cítrico) no vinho; <BR>
<P align = "justify"> 2 - volatile acidity (acidez volátil) - medida da acidez devido a ácidos de baixo peso molecular (sobretudo ácido acético) presentes no vinho, que são responsáveis pelo aroma e gosto de vinagre; <BR>
<P align = "justify"> 3 - citric acid (ácido cítrico) - medida de ácido cítrico no vinho;<BR>
<P align = "justify"> 4 - residual sugar (açúcar residual) - medida de açúcar residual presente no vinho, com origem nos resíduos de açúcar da uva que permanecem no vinho após o fim da fermentação;<BR>
<P align = "justify"> 5 - chlorides (cloretos) - medida de cloretos (íons de cloro) no vinho; <BR>
<P align = "justify"> 6 - free sulfur dioxide (dióxido de enxofre livre)- medida de dióxido de enxofre livre (isto é, que não está ligado a outras moléculas) no vinho;<BR>
<P align = "justify"> 7 - total sulfur dioxide - (dióxido de enxofre total) - medida de dióxido de enxofre total (livre + porção ligada a outras moléculas) no vinho;<BR>
<P align = "justify"> 8 - density (densidade) - medida da densidade do vinho;<BR>
<P align = "justify"> 9 - pH - medida do pH do vinho;<BR>
<P align = "justify"> 10 - sulphates (sulfatos) - medida de sulfatos (íons SO₄²⁻) no vinho;<BR>
<P align = "justify"> 11 - alcohol (teor alcoólico)- medida da graduação alcoólica do vinho.<BR>
<P align = "justify"> 12 - quality (qualidade) -  score numérico de qualidade (de 0 a 10), produzido com base em dados sensoriais, sendo 0 (muito ruim) e 10 (excelente). </P>

## Visualização da distribuição dos dados 

```python
#observa-se uma distribuição normal dos dados

for col in df:
    
    sns.histplot(data=df, x=col, kde=True).set_title(f"Distribuição da variável {col}")
    plt.show()
```

<b> Verificação de outliers com o uso do z score > 3 </b>

```python
for col in df:

    mu, std = df[col].mean(), df[col].std() 
    aux_outlier = df[col].apply(lambda x: (x-mu)/std).apply(lambda x: True if abs(x) > 3 else False)
    
    if aux_outlier.sum() > 0:
        
        print(f"Outliers de {col}:\n")
        
        print(aux_outlier[aux_outlier].index.tolist())
        
    else:
        print(f"Não há outliers na coluna {col}!")
        
    print()
    print("="*80)
    print()
```

```python
# Em análise sensorial, uma escala de 10 pontos, uma nota em escalas abaixo de 5 são consideradas ruins

df["quality_bin"] = df["quality"].apply(lambda x: "good" if x>5 else "bad")

df["quality_bin"].value_counts(normalize=True)
```

```python
# Verificar nota de corte para atribuir a qualidade do vinho
# Observa-se que a nota de corte que melhor minimiza o desbalanceamento das classes, é a nota 5

for corte in df["quality"].sort_values().unique():
    
    print(f"\nDistribuição de classes para corte em {corte}:")
    
    aux_bin = df["quality"].apply(lambda x: 1 if x > corte else 0)
    
    display(aux_bin.value_counts(normalize=True))
    
    sns.countplot(x=aux_bin)
    plt.show()
```

```python
df["quality_bin"] = df["quality"].apply(lambda x: "good" if x>5 else "bad")
```

```python
df = df.drop(columns="quality")
```

```python
df.to_csv("winequality-red-binary.csv", index=False)
```

```python
#Considerando agora, a base de vinhos tintos com a variável quality discretizada em duas classes 
#("good" para score maior que 5; "bad" caso contrário). 
```

<b>Avaliação de Modelo (Teste e Treino)</b>

```python
from sklearn.model_selection import train_test_split

X = df.drop(columns=["quality_bin"])
y = df["quality_bin"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, stratify=y, random_state=42)

df_train = pd.concat([X_train, y_train], axis=1)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
```

```python
#uso de regressão logística e regularização dos dados

etapas = [("escaler", StandardScaler()), 
          ("logit", LogisticRegression())]

Pipeline(etapas)
```

```python
pile_logit = Pipeline(etapas).fit (X_train, y_train)

pile_logit
```

```python
etapas = [("escaler", StandardScaler()), 
          ("logit", LogisticRegression(max_iter=1000))]
```

```python
pile_logit = Pipeline(etapas).fit (X_train, y_train)

pile_logit
```

```python
# Predição

y_pred = pile_logit.predict (X_test)
```

```python
# Métricas de avaliação para o meu modelo

from sklearn.metrics import ConfusionMatrixDisplay, classification_report
```

O modelo predito foi capaz de acertar 191 casos em que o vinho pode ser considerado de qualidade boa,
e 161 vinhos em que eles eram considerados ruins, no entanto, o modelo apresentou erro de 66 vinhos que deveriam ser considerados de qualidade boa, mas foram classificados como ruins.

```python
ConfusionMatrixDisplay.from_predictions (y_test, y_pred);
```

```python
print (classification_report (y_test, y_pred))
```
